```python
def find_citates_ind(original_text, target_text):
    
    target_text = target_text.sents
    original_text = original_text.sents
    
    #print(list(target_text))
    #print(list(original_text))
    
    citates_ind = list()
    
    for targ_ind, targ_sent in enumerate(target_text):
        targ_sent = re.split(r'[.,;\s]', str(targ_sent))
        
        for orig_ind, orig_sent in enumerate(original_text):                        
            orig_sent = re.split(r'[.,;\s]', str(orig_sent))
            
            if len(targ_sent) == len(orig_sent):
                citate_flag = True
                
                for targ_word, orig_word in zip(targ_sent, orig_sent):
                    if targ_word != orig_word:
                        citate_flag = False
                        break
                
                if citate_flag:
                    citates_ind.append(
                    {
                        "targ_sent_ind": targ_ind,
                        "orig_sent_ind": orig_ind,
                    })                

    return citates_ind
```
